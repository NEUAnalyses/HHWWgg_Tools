```python
import uproot
import os 
from matplotlib import pyplot as plt 
import numpy as np 
```

```python
direc = "/eos/user/b/bmarzocc/HHWWgg/January_2021_Production/2017/Backgrounds/"
ol = "/eos/user/a/atishelm/www/HHWWgg/DNN/kinWeights/"

WeightTypes = ["kinWeight.", "kinWeight_noHgg"]
WeightTypesDict = {
    "kinWeight." : "WithHgg",
    "kinWeight_noHgg" : "NoHgg"
}

for WT in WeightTypes:
    files = ["%s/%s"%(direc, f) for f in os.listdir(direc) if WT in f]
    weightTypeLabel = WeightTypesDict[WT]
    print("Weight Type:",weightTypeLabel)
    global_min = 1000000,
    global_max = -1000000
    Wtk_integrals = []
    for file in files:
        print("on File:",file)
        fName = file.split('/')[-1]
        f = uproot.open(file) 
        treeName_bytes = f.keys()[0]
        treeName = treeName_bytes.decode("utf-8")
        tree = f[treeName]

        kinWeight = tree['kinWeight'].array()
        weight = tree['weight'].array()
        kinTimesWeight = np.multiply(kinWeight, weight)

#         minVal = min(kinWeight)
#         maxVal = max(kinWeight)
#         print("min kinWeight:",minVal)
#         print("max kinWeight:",maxVal)

#     #     bins = np.linspace(0, 20000000, 20)
#     #     fig, ax = plt.subplots()
#     #     plt.hist(kinWeight, bins = bins, label = "kinWeight")
#     #     plt.yscale('log')
#     #     plt.savefig("%s/%s_kinWeights.png"%(ol, fName))
#     #     plt.close()

#         minVal = min(weight)
#         maxVal = max(weight)
#         print("min weight:",minVal)
#         print("max weight:",maxVal)    

#     #     bins = np.linspace(-20000000, 20000000, 40)
#     #     fig, ax = plt.subplots()
#     #     plt.hist(weight, bins = bins, label = "weight")
#     #     plt.yscale('log')
#     #     plt.savefig("%s/%s_weight.png"%(ol, fName))
#     #     plt.close()  


        print('----')
        minVal = min(kinTimesWeight)
        maxVal = max(kinTimesWeight)    
        mean = np.average(kinTimesWeight)
        std_val = np.std(kinTimesWeight)
        print("min kinTimesWeight:",minVal)
        print("max kinTimesWeight:",maxVal)
        print("average kinTimesWeight:",np.average(kinTimesWeight))
        print("average kinTimesWeight:",np.std(kinTimesWeight))
        
        if(minVal < global_min):
            global_min = minVal
        if(maxVal > global_max):
            global_max = maxVal            
        
        
        
        KTW_sum = sum(kinTimesWeight)
        print("kinTimesWeight Integral:",KTW_sum)
        Wtk_integrals.append(KTW_sum)
        
        bins = np.linspace(-50000, 250000, 200)
        fig, ax = plt.subplots()
#         plt.hist(kinWeight, bins = bins, label = "kinWeight")
#         plt.hist(weight, bins = bins, label = "weight")
        plt.hist(kinTimesWeight, bins = bins, label = "Weight * kinWeight")
        plt.yscale('log')
        
        textstr = '\n'.join((
            r'$\mu=%.4f$' % (mean, ),
            r'$\sigma=%.4f$' % (std_val, )))   
        props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
        ax.text(0.4, 0.85, textstr, transform=ax.transAxes, fontsize=14,
                verticalalignment='top', bbox=props)  
        
        plt.legend()
        plt.savefig("%s/%s/%s_WeightTimeskinWeight.png"%(ol, weightTypeLabel, fName))
        plt.savefig("%s/%s/%s_WeightTimeskinWeight.pdf"%(ol, weightTypeLabel, fName))
        plt.close()    
        
        ##-- Apply cut of +/-5 on weightTimesKin
        Nbefore = len(kinTimesWeight)
        kinTimesWeight = kinTimesWeight[abs(kinTimesWeight) < 10]
        Nafter = len(kinTimesWeight)
        
        NCut = int(Nbefore) - int(Nafter)
        Ratio_Cut = round(1. - (float(Nafter)/float(Nbefore)), 4)
        
        print("Number of events cut:",NCut)
        print("Ratio Cut:",Ratio_Cut)
        
        ##-- Recompute mean and stdev 
        mean = np.average(kinTimesWeight)
        std_val = np.std(kinTimesWeight)
        
        bins = np.linspace(-50000, 250000, 200)
        fig, ax = plt.subplots()
#         plt.hist(kinWeight, bins = bins, label = "kinWeight")
#         plt.hist(weight, bins = bins, label = "weight")
        plt.hist(kinTimesWeight, bins = bins, label = "Weight * kinWeight : abs() < 10")
        plt.yscale('log')
        
        textstr = '\n'.join((
            r'$\mu=%.4f$' % (mean, ),
            r'$\sigma=%.4f$' % (std_val, ),
            "Ratio Cut: %s"%(Ratio_Cut)
        ))   
        props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
        ax.text(0.4, 0.85, textstr, transform=ax.transAxes, fontsize=14,
                verticalalignment='top', bbox=props)  
        
        plt.legend()
        plt.savefig("%s/%s/%s_WeightTimeskinWeight_withCut.png"%(ol, weightTypeLabel, fName))
        plt.savefig("%s/%s/%s_WeightTimeskinWeight_withCut.pdf"%(ol, weightTypeLabel, fName))
        plt.close()          

    print("Weight Type:",weightTypeLabel)
    print("Global Min:",global_min)
    print("Global Max:",global_max)
    print("Wtk_integrals:",Wtk_integrals)
    
print("DONE")
```
