```python

def GetSystLabels(year_):
    systLabels = [""]    

    
    listOfSources_2016 = [
                                "Absolute",
                                "Absolute2016",
                                "BBEC1",
                                "BBEC12016",
                                "EC2",
                                "EC22016",
                                "FlavorQCD",
                                "HF",
                                "HF2016",
                                "RelativeBal",
                                "RelativeSample2016"
                               ]

    listOfSources_2017 = [
                        "Absolute",
                        "Absolute2017",
                        "BBEC1",
                        "BBEC12017",
                        "EC2",
                        "EC22017",
                        "FlavorQCD",
                        "HF",
                        "HF2017",
                        "RelativeBal",
                        "RelativeSample2017"
                        ]

    listOfSources_2018 = [
                        "Absolute",
                        "Absolute2018",
                        "BBEC1",
                        "BBEC12018",
                        "EC2",
                        "EC22018",
                        "FlavorQCD",
                        "HF",
                        "HF2018",
                        "RelativeBal",
                        "RelativeSample2018"
                        ]  

    

    ##-- Systematic trees dictionary 
    for direction in ["Up","Down"]:

        ##-- Photons 
        systLabels.append("MvaShift%s01sigma"%direction)
        systLabels.append("SigmaEOverEShift%s01sigma"%direction)
        systLabels.append("MaterialCentralBarrel%s01sigma"%direction)
        systLabels.append("MaterialOuterBarrel%s01sigma"%direction)
        systLabels.append("MaterialForward%s01sigma"%direction)
        systLabels.append("FNUFEB%s01sigma"%direction)
        systLabels.append("FNUFEE%s01sigma"%direction)
        systLabels.append("MCScaleGain6EB%s01sigma"%direction)
        systLabels.append("MCScaleGain1EB%s01sigma"%direction)

        ##-- Jets 
        systLabels.append("JEC%s01sigma" % direction)
        systLabels.append("JER%s01sigma" % direction)           
        systLabels.append("PUJIDShift%s01sigma" % direction)

        ##-- Different reduced JEC names for different years
        if(year == '2016'):
            for sourceName in listOfSources_2016:
                systLabels.append("JEC%s%s01sigma" % (str(sourceName),direction))
        elif(year == '2017'):
            for sourceName in listOfSources_2017:
                systLabels.append("JEC%s%s01sigma" % (str(sourceName),direction))  
        elif(year == '2018'):
            for sourceName in listOfSources_2018:
                systLabels.append("JEC%s%s01sigma" % (str(sourceName),direction))  

        ##-- HEM Systematic only in 2018 
        if(year == '2018'):
            systLabels.append("JetHEM%s01sigma" % (direction))

        ##-- MET 
        systLabels.append("metJecUncertainty%s01sigma" % direction)
        systLabels.append("metJerUncertainty%s01sigma" % direction)
        systLabels.append("metPhoUncertainty%s01sigma" % direction)
        systLabels.append("metUncUncertainty%s01sigma" % direction)        

        for r9 in ["HighR9","LowR9"]:
            for region in ["EB","EE"]:
                systLabels.append("ShowerShape%s%s%s01sigma"%(r9,region,direction))
                systLabels.append("MCScale%s%s%s01sigma" % (r9,region,direction))
                for var in ["Rho","Phi"]:
                    systLabels.append("MCSmear%s%s%s%s01sigma" % (r9,region,var,direction))

    return systLabels
                        
    
print("DONE")
#     print("Number of systLabels:",len(systLabels))
#     print("systLabels:",systLabels)
```

```python
##-- Want a scale factor for each (year, proc, cat) 
##-- ggh_125_13TeV_HHWWggTag_SL_0;1

##-- All 
# years = ["2016", "2017", "2018"]
# procs = ["ggh", "vbf", "wzh", "tth"]
# cats = ["SL_0", "SL_1", "SL_2", "SL_3"]

# ##-- Less
# years = ["2016"]
# procs = ["wzh", "tth"]
# cats = ["SL_0"]

##-- Custom
years = ["2016", "2017", "2018"]
# procs = ["wzh", "tth"]
procs = ["ggh", "vbf"]
cats = ["SL_0", "SL_1", "SL_2", "SL_3"]


maxSystLabels_ = -1 ##-- -1 to run over all systematic labels
maxSystLabels = int(maxSystLabels_)

BaseDirec = "/eos/user/a/atishelm/ntuples/HHWWgg_flashgg/January_2021_Production/"
TrainingLabel = "HHWWyyDNN_WithHggFactor2-200Epochs-3ClassMulticlass_EvenSingleH_2Hgg_withKinWeightCut10_BalanceYields" ##-- AN_20_165_v6
# TrainingLabel = "HHWWyyDNN_MultiClass_2Hgg_HggClassWeightFactor2_BalanceYields"

Single_H_Info_Dict = {
    "ggh" : "GluGluHToGG",
    "vbf" : "VBFHToGG", 
    "wzh" : "VHToGG",
    "tth" : "ttHJetToGG"
}



txtFileName = "Single_Higgs_ScaleFactors_%s.txt"%(TrainingLabel)
os.system("rm %s"%(txtFileName))
outFile = open(txtFileName, "w")
outFile.write("year proc cat systematic AllOverOddSF\n")

for year in years:
    print("year:",year)
    
    systematicLabels = GetSystLabels(year)
    
    for proc in procs:
        print("proc:",proc)
        for cat in cats:
            print("cat:",cat)
            SHlabel = Single_H_Info_Dict[proc]
#             tree = "%s_125_13TeV_HHWWggTag_%s"%(proc, cat)

            scaleFactors = []
            ##-- For each systematic label 
            for isyst, systLabel in enumerate(systematicLabels):
                print("isyst:",isyst)
                if(isyst == maxSystLabels): 
                    print("Reached manual maximum of %s systematic labels"%(maxSystLabels))
                    print("Manual exit")
                    break 
                
                ##-- Different tree name for each systematic label 
                systExt= "_%s_v1"%(systLabel)
                tree = "NOTREE"
                
                if(systLabel == ""):
                    print("nominal tree")
                    tree = "%s_125_13TeV_HHWWggTag_%s"%(proc, cat)
                else:
                    print("systematic:",systLabel)
#                     systExt= "_%s_v1"%(systLabel)
#                     thisExt = cat.replace(, systExt)
                    tree = "%s_125_13TeV_HHWWggTag_%s_%s"%(proc, cat, systLabel)
                    
                print("tree:",tree)
                
#                 ggh_125_13TeV_HHWWggTag_0_metPhoUncertaintyDown01sigma_v1
#                 tree = "%s_125_13TeV_HHWWggTag_%s"%(proc, cat)

                ##-- Full events file 
                AllEventsFile = "%s/%s/%s_FullHiggsSamplesOnly_v2/Single_H/SingleHiggs_%s_%s_all_CategorizedTrees.root"%(BaseDirec, year, TrainingLabel, SHlabel, year)
                AEF = uproot.open(AllEventsFile)
                AllEventsWeights = AEF[tree]["weight"].array()
                AllEventsWeightsSum = np.sum(AllEventsWeights)

                ##-- Odd events file 
                OddEventsFile = "%s/%s/%s/Single_H/SingleHiggs_%s_%s_all_CategorizedTrees.root"%(BaseDirec, year, TrainingLabel, SHlabel, year)
                HEF = uproot.open(OddEventsFile)
                HalfEventsWeights = HEF[tree]["weight"].array()
                HalfEventsWeightsSum = np.sum(HalfEventsWeights)

                ##-- Compute scale factor as ratio of two weighted yields 
                SF = float(AllEventsWeightsSum) / float(HalfEventsWeightsSum)

                ##-- Save SF as line in output text file 
                if(systLabel == ""):
                    systLabel = "NOMINAL"
                print("SF:",SF)
                line = "%s %s %s %s %s\n"%(year, proc, cat, systLabel, str(SF))
                outFile.write(line)
                scaleFactors.append(SF)
                
#             print("scaleFactors:",scaleFactors)
            scaleFactors = np.array(scaleFactors)
            minValue = min(scaleFactors)
            maxValue = max(scaleFactors)
            mean = scaleFactors.mean()
            stdev = scaleFactors.std()
            
            print("mean:",mean)
            print("stdev:",stdev)
            
            fig, ax = plt.subplots()
            xmin, xmax, nbins = 0, 5, 51
            bins = np.linspace(xmin, xmax, nbins + 1)
            label = "%s_%s_%s"%(year, proc, cat)
            plt.hist(scaleFactors, bins = bins, label = label)
            plt.xlabel("Scale factor")
            plt.ylabel("Entries")
            plt.xlim(minValue - 1, maxValue + 1)
            plt.legend()
            plt.grid()
            
            textstr = '\n'.join((
                r'$\mu=%.4f$' % (mean, ),
                r'$\sigma=%.4f$' % (stdev, )))     
            props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
            infoBox_xPosition, infoBox_yPosition = 0.1, 0.9
            ax.text(infoBox_xPosition, infoBox_yPosition, textstr, transform=ax.transAxes, fontsize=14,
                    verticalalignment='top', bbox=props)              
            
#             plt.show()
            plt.title("%s Semi-Leptonic Systematic Scale Factors"%(label))
            plt.savefig("/eos/user/a/atishelm/www/HHWWgg/Post_AN_20_165_v6/%s.png"%(label))
            plt.savefig("/eos/user/a/atishelm/www/HHWWgg/Post_AN_20_165_v6/%s.pdf"%(label))
            plt.close()
            

outFile.close()
print("DONE")            
```
