```python
def GetEvents(mass_):
    evDict = {
        "X250": 98443,
        "X260": 99932,
        "X270": 99952,
        "X280": 99936,
        "X300": 99931,
        "X320": 99941,
        "X350": 99945,
        "X400": 99937,
        "X500": 99947,
        "X550": 99940,
        "X600": 99961,
        "X650": 99450,
        "X700": 99449,
        "X750": 99953,
        "X800": 99443,
        "X850": 99942,
        "X900": 99945,
        "X1000": 99450,
        "X1250": 97940,
        "SM": 99931,
        "node2": 99944,
        "node9": 99938
    }

    return evDict[mass_]

def Draw_Histogram(h_,opt_,fn_):
    c_tmp = ROOT.TCanvas('c_tmp','c_tmp',1300,800)
    h_.Draw(opt_)
    c_tmp.SaveAs(fn_)
    
def Draw_Multigraph(h_,opt_,fn_):
    c_tmp = ROOT.TCanvas('c_tmp','c_tmp',1300,800)

    # if log_:     
        # print'h_ = ',h_
    # axis = h_.GetYaxis()
    # axis.SetLimits(0.00001,1)
    # h_.GetYaxis().SetLimits(0.00001,1)
    h_.Draw("apl")
#     if(log_): gPad.SetLogy()

    # else:


#     h_.Draw(opt_)
        # c_tmp.BuildLegend()
        # c_tmp.BuildLegend(0.65,0.52,0.95,0.82)

    # c_tmp.BuildLegend(0.75,0.62,0.95,0.84)
    c_tmp.BuildLegend(0.75,0.42,0.95,0.65)

    # c_tmp.BuildLegend(0.83, 0.83, 0.99, 0.99)
    # c_tmp.BuildLegend(0.5, 0.3, 0.8, 0.8)
    # c_tmp.BuildLegend(0.6, 0.1, 0.9, 0.6)
    c_tmp.SaveAs(fn_)
#     h_.SetName('h')
#     h_.SaveAs("h.root")
    return 

def SetGraphStyle(gr_,num_,clr_,msty_): 
    gr_.SetLineStyle(num_)
    gr_.SetMarkerStyle(msty_)
    gr_.SetLineColor(clr_)
    gr_.SetMarkerColor(clr_)
    gr_.SetMarkerSize(1.5)
    return 0 
```
